```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

# Plot the original waypoints

```python
df = pd.read_csv('data/highway_map.csv', sep=' ', header=None)
df.columns = ['x','y','s','d_x','d_y']
```

```python
df.head(10)
```

```python
# check that frenet d vectors are unit lenght
max(df.d_x*df.d_x+df.d_y*df.d_y)
```

```python
r = [x for x in range(len(df))]
plt.scatter(df.x[r], df.y[r], c=df.s[r]);
plt.colorbar();
# colorbar represents S coordinate
```
